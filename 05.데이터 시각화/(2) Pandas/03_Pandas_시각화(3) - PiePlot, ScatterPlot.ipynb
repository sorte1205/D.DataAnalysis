# Pandas 시각화 - PiePlot, ScatterPlot

DataFrame은 내부적으로 `pyplot` 객체를 초기화고 있기 때문에 이를 활용하여 그래프를 직접 표시할 수 있다.

## #01. 준비작업

### [1] 패키지 참조

```python
import sys
import numpy as np
from pandas import read_excel
from matplotlib import pyplot as plt
```

### [2] 그래프 전역 설정

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 9
plt.rcParams["figure.figsize"] = (15, 6)
plt.rcParams['figure.dpi'] = 200
plt.rcParams["axes.unicode_minus"] = False
```

## #02. 파이 그래프

### [1] 데이터 가져오기

#### (1) 데이터프레임 생성

```python
df1 = read_excel("https://data.hossam.kr/pydata/city_people.xlsx")
df1
```

#### (2) 데이터 전처리

도시별 평균 인구수를 구한다.

```python
df1_mean = df1[['도시', '인구']].groupby('도시').sum()
df1_mean
```

### [2] 파이 그래프 시각화

파이 그래프의 경우 반드시 데이터 프레임 안에서 특정 변수를 지정해야 한다.

데이터 프레임의 index가 파이 그래프의 각 조각이 된다.

```python
df1_mean['인구'].plot.pie(autopct='%0.1f%%')
plt.show()
plt.close()
```

## #03. 산점도 그래프

### [1] 데이터 가져오기

```python
df2 = read_excel("https://data.hossam.kr/pydata/traffic_acc.xlsx")
df2
```

### [2] 추세선을 위한 회귀식 구하기

```python
x = df2['발생건수']
z = np.polyfit(x, df2['부상자수'], 1)
f = np.poly1d(z)
y = f(x)
y
```

### [3] 산점도 그래프 시각화

산점도 그래프는 데이터프레임에 대해 $x$축으로 사용될 변수와 $y$축으로 사용될 변수를 설정해야 한다.

추세선은 `pyplot` 객체를 활용하여 직접 구현해야 한다.

```python
df2.plot.scatter(x='발생건수', y='부상자수', color="#0066ff")
plt.plot(x, y, color="#ff0000", linewidth=1)

plt.grid()

plt.show()

plt.close()
```
