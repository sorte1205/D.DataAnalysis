# 연구과제(Pie, Scatter) 풀이

## 준비과정

### [1] 패키지 참조

```python
import sys
import numpy as np
from pandas import read_excel, DataFrame
from matplotlib import pyplot as plt
```

### [2] 그래프 초기화

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.unicode_minus"] = False
```

## 문제 1

```python
df1 = read_excel("https://data.hossam.kr/pydata/poll.xlsx")
df1
```

```python
df1['연령대'] = df1['나이'].apply(lambda x: "%d대" % ((x//10) * 10))
df1
```

```python
ratio = df1['연령대'].value_counts()
ratio
```

```python
plt.figure()
plt.title("연령대 비율")
plt.pie(ratio.values, labels=ratio.index, autopct='%0.1f%%')

txt = """총 40명의 조사 인원 중에서 20대의 비율이 42.5%로 가장 많았다.
그 다음으로는 30대가 25%, 40대가 20%, 50대가 12.5%의 순으로 나타났다."""
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```

## 문제 2

```python
df2 = read_excel("https://data.hossam.kr/pydata/poll.xlsx")
df2
```

```python
conditions = [(df2['월수입'] < 100),
              (df2['월수입'] < 250)]

value = ['저소득', '보통']
df2['소득수준'] = np.select(conditions, value, default='고소득')
df2
```

```python
ratio = df2['소득수준'].value_counts()
ratio
```

```python
plt.figure()
plt.title("소득수준 비율")
plt.pie(ratio.values, labels=ratio.index, autopct='%0.1f%%')

txt = """총 40명의 조사 인원 중에서 보통의 소득 수준을 갖는 경우는 55%로 절반 이상을 차지했다.
고소득은 7.5%에 불과했으며 저소득 비율은 37.5%로 집계되었다."""
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```

## 문제 3

```python
df3 = read_excel("https://data.hossam.kr/pydata/poll.xlsx")
df3
```

```python
z = np.polyfit(df3['나이'], df3['월수입'], 1)
f = np.poly1d(z)
f
```

```python
x = df3['나이']
y = f(df3['나이'])
y
```

```python
plt.figure()
plt.scatter(df3['나이'], df3['월수입'])
plt.plot(x, y, color="#ff0000", linewidth=0.5, linestyle=":")
plt.title("나이에 따른 소득")
plt.xlabel("나이")
plt.ylabel("소득")
plt.grid()

txt = "f(x) = %f * x + %f\n\n" % (z[0], z[1])
txt += "조사 대상의 나이와 소득은 양의 상관관계가 있는 것으로 나타났다."
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```

```python
x = [20, 25, 30, 35, 40, 45, 50]
y = f(x)
y
```

```python
plt.figure()
plt.title("연령별 예상 월소득")
plt.bar(x, y, width=2.5)
plt.xlabel("연령")
plt.ylabel("월소득")
plt.xticks(x, x)
plt.grid()

txt = "조사 대상의 연령이 높아질 수록 월 소득 수준도 함께 높아진다."
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```

## 문제 4

```python
df4 = read_excel("https://data.hossam.kr/pydata/jobs_by_scale.xlsx", index_col='종사자규모별')
df4
```

```python
mean = df4.mean(axis=1)
mean
```

```python
plt.figure()
plt.title("종사자 규모별 업장 수 비율")
plt.pie(mean.values, labels=mean.index, autopct='%0.1f%%')

txt = """종사자 수가 1~4명인 업장이 24.4%로 가장 높은 비율을 나타냈으며
1000명 이상이 종사하는 대규모 업장이 23.9%로 높게 나타났다.
그 밖의 종사자 규모는 모두 10% 미만인 것으로 나타났다.
1000명 이상의 대규모 보다는 1000명 미만의 중소기업이 대다수를 차지하고 있는 것으로 볼 수 있다."""
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```

## 문제 5

```python
df5 = read_excel("https://data.hossam.kr/pydata/traffic_acc.xlsx")
df5
```

```python
z = np.polyfit(df5['발생건수'], df5['부상자수'], 1)
print("기울기(가중치) =", z[0])
print("절편(편향) = ", z[1])
```

```python
f = np.poly1d(z)
f
```

```python
x = df5['발생건수']
y = f(x)
y
```

```python
plt.figure()
plt.scatter(df5['발생건수'], df5['부상자수'])
plt.plot(x, y, color="#ff0000", linewidth=0.5, linestyle=":")
plt.title("교통사고 발생건수에 따른 부상자수")
plt.xlabel("발생건수")
plt.ylabel("부상자수")
plt.grid()

txt = "f(x) = %f * x + %f\n\n" % (z[0], z[1])
txt += "교통사고 발생건수가 증가할 수록 부상자수 역시 증가하는 것으로 보인다."
plt.figtext(0.51, 0, txt, wrap=True,
            verticalalignment="top", horizontalalignment="center",
            fontdict={"size": 11, "weight": "light"})

plt.show()
plt.close()
```
