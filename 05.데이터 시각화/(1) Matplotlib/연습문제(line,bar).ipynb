문제 1

다음 데이터는 2020년 2월 5일부터 2023년 5월 31일까지 서울과 전국의 Covid19 일일 확진자 수를 기록한 데
이터이다. 조사 기간동안 서울과 전국의 확진자 수가 어떻게 변화하고 있는지에 대한 추이를 시각화 하라.

```python
import numpy as np
from pandas import read_excel, DataFrame, to_datetime
from matplotlib import pyplot as plt
from matplotlib import font_manager
```

```python
origin = read_excel("https://data.hossam.kr/pydata/covid19_active.xlsx")
origin
```

```python
plt.rcParams["font.family"]='Malgun Gothic'
plt.rcParams["font.size"] =12
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams['axes.unicode_minus'] = False
```

```python
df= origin
```

```python
y1 = df['서울시 일일 확진']
y2 = df['전국 일일 확진']
x = to_datetime(df['기준일'])
```

```python
plt.figure()   # 그래프 시작을 선언
plt.title("일일 확진자 추이")          # 그래프 제목 표시
plt.plot(x, y1, color='#74BDCB', label='서울시')  
plt.plot(x, y2, color='#FFA384', label='전국')  
plt.ylabel("확진자 수")               # y축 제목
plt.legend()
plt.grid()                    # 배경 격자 표시
plt.show()                    # 그래프 화면 표시
plt.close()                   # 그래프 종료
```

문제4

다음 데이터는 도시별 인수수를 5개년도 단위로 조사한 가상의 자료이다.
각 도시별로 2015년도의 2010년도의 인구수를 비교하는 자료를 시각화 하시오.
단, 인구수의 표현 범위는 200만~1000만으로 한다

```python
origin = read_excel("https://data.hossam.kr/pydata/city_people.xlsx")
origin
```

```python
from pandas import read_excel, pivot_table, melt
```

```python
df4 = pivot_table(origin,
            index = '도시',
            columns = '연도',
            values = '인구',
)
df4.dropna(inplace = True, axis = 1)
df4 = df4.reset_index()
df4
```

```python
x = np.arange(0, len(df4['도시']))
y1 = df4[2010]
y2 = df4[2015]
xticks = []

for i in df4['도시']:
    xticks.append("%s" % i)

x, y1, y2, xticks
```

```python
plt.figure()
plt.title("도시별 인구 수")
plt.bar(x-0.2, y1, color='#74BDCB', label='2010', width=0.4)  
plt.bar(x+0.2, y2, color='#FFA384', label='2015', width=0.4)  
plt.ylabel("인구 수")
plt.xlabel("도시")
plt.xticks(x, xticks)
plt.legend()
plt.grid()
plt.show()
plt.close()
```

```python
plt.rcParams["font.family"]='Malgun Gothic'
plt.rcParams["font.size"] =12
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams['axes.unicode_minus'] = False
```

종사자 규모에 따른 업체 수를 확인할 수 있는 세로 막대 그래프를 구현하시오. 이 때 각 데이터의 범주는 년도
로 설정하시오.

```python
origin1 = read_excel("https://data.hossam.kr/pydata/jobs_by_scale.xlsx")
origin1
```

```python
y1 = origin1[2015]
y2 = origin1[2016]
# float(y1.values)
# float(y2.values)
x = np.arange(0,len(origin1['종사자규모별']))

xticks = []
for i in origin1['종사자규모별']:
    xticks.append("%s" % i)
x, y1, y2
```
