```python
# 최소
df2 = origin.copy()
df3['최소'] = df3.iloc[:,2:6].min(axis=1)
df3
```

```python
# 표준편차
df2 = origin.copy()
df3['표준편차'] =df3.iloc[:,2:6].std(axis=1)
df3
```

```python
# 분산
df3['분산']=df3.iloc[:,2:6].var(axis=1)
df3
```

### [3] 조건별 파생변수

#### (1) if ~ else

합계와 평균에 대한 파생변수 추가

```python
df4 = origin.copy()
df4['합계'] = df4.iloc[:,2:6].sum(axis=1)
df4['평균'] = df4.iloc[:,2:6].mean(axis=1)
df4
```

numpy를 사용한 평균값에 따른 조건부 파생변수

```python
df4['결과']=np.where(df4['평균']>=80, '합격', '불합격')  #조건에 부합하면 '합격', 아니면 '불합격'
df4
```

표현식을 사용한 평균값에 따른 조건부 파생변수

```python
df4['결과2'] ="불합격"
df4.loc[df4['평균']>=80, "결과2"] = '합격'
df4
```

#### (2) 경우의 수에 따른 조건값 (elif)

numpy 사용

```python
# 학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [  (df4['평균'] >= 90),   # if
                (df4['평균'] >= 80),   # elif
                (df4['평균'] >= 70)    # elif
]

# 조건에 따라 부여될 학점
value = ['A', 'B', 'C']

#                        조건        값        else
df4['학점1'] = np.select(conditions, value, default='F')
df4
```

사용자 정의 함수 사용

```python
# 학점을 부여하기 위한 점수의 구간을 함수로 지정
def getScore(x):
    if x >= 90:
        score = 'A'
    elif x >= 80:
        score = 'B'
    elif x >= 70:
        score = 'C'
    else:
        score ='F'
    return score
```

```python
df4['학점2'] = df4['평균'].apply(getScore)

```
